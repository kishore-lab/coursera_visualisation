```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("coursea_data.csv")
```

```
df.head()
```

```
labels=['COURSE','SPECIALIZATION','PROFESSIONAL CERTIFICATE']
data=df.course_Certificate_type.value_counts()
print(data)
```

```

plt.figure(figsize=(20,6))
plt.pie(data,labels=labels,explode=[0.02,0.02,0.02],autopct='%0.02f%%')
plt.title("Course certification Analysis",fontsize=16)

plt.legend()
plt.show()
```

```
data=df.course_difficulty.value_counts()
print(data)
labels=['Beginner','Intermediate','Mixed','Advanced']
plt.figure(figsize=(20,6))
plt.pie(data,labels=labels,autopct='%0.02f%%',explode=[0.01,0.01,0.01,0.01])
plt.title("Course difficulty Analysis",fontsize=16)

plt.legend()
plt.show()
```

```
plt.figure(figsize=(10,7))
plt.title("course Certificate Analysis")

ax=sns.countplot(x="course_Certificate_type",data=df)

for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d'% int(p.get_height()),
           fontsize=12,color='blue',ha='center',va='bottom')
```

```
plt.figure(figsize=(10,7))
plt.title("course difficulty Analysis")

ax=sns.countplot(x="course_difficulty",data=df)

for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d'% int(p.get_height()),
           fontsize=12,color='brown',ha='center',va='bottom')
```

```
sns.catplot(x='course_Certificate_type',y='course_rating',kind='bar',data=df,ci=None,palette='husl')
plt.ylabel('course rating',fontsize=14)
plt.ylim(4.4,4.8)
plt.title('Relation between type, difficulty and rating')
```
